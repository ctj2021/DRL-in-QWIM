```python
import yfinance as yf
import pandas as pd
import numpy as np
import pickle
```

a description of world indices symbol-name pair   
https://finance.yahoo.com/world-indices 

```python
index_list = ['^GSPC', '^IXIC', '^RUT']
sp500 = yf.download(index_list[0], start="2000-01-01", end="2022-06-30")
nasdaq = yf.download(index_list[1], start="2000-01-01", end="2022-06-30")
russell2000 = yf.download(index_list[2], start="2000-01-01", end="2022-06-30")
```

```python
sp500
```

```python
with open('data/sp500.pkl', 'wb') as f:
    pickle.dump(sp500, f)

with open('data/nasdaq.pkl', 'wb') as f:
    pickle.dump(nasdaq, f)

with open('data/russell2k.pkl', 'wb') as f:
    pickle.dump(russell2000, f)
```
